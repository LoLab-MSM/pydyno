```python
from tropical.cytoscapejs_visualization.model_visualization import ModelVisualization
from tropical.examples.EARM.earm2_flat import model
import pickle
# Need to import the clus_info to visualize
with open('results_spectral.pickle', 'rb') as handle:
    clus_info = pickle.load(handle)
```

```python
set(clus_info[13]["labels"])
```

```python
import numpy as np
# Loading dynamic signatures
with open('earm_signatures.pickle', 'rb') as handle:
    earm_signatures = pickle.load(handle)
# Obtaining signature of how the species 0 (the enzyme) is being consumend
sp37_sign_reactants = earm_signatures[37]['consumption']

# Initializing clustering object
clus = clustering.ClusterSequences(sp37_sign_reactants, unique_sequences=False, truncate_seq=50)
# clus.diss_matrix()
diss = np.load('earm_sp37_diss.npy')
clus.diss = diss
```

```python
earm_signatures['species_combinations'][37]['reactants']
```

```python
# clustering signatures
clus.spectral_clustering(n_clusters=11, n_jobs=4, random_state=1234)
```

```python
clus.labels[0]
```

```python
%matplotlib inline

pl37 = clustering.PlotSequences(clus)
pl37.legend_plot(model=model, reactions_comb=earm_signatures['species_combinations'][37]['reactants'], combs=[14, 4, 20, 16, 44, 5, 86, 15, 45, 50, 1])
```
