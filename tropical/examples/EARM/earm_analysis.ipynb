```python
# Importing libraries
import pickle
from tropical import clustering
```

```python
# Loading dynamic signatures
with open('results3.pickle', 'rb') as handle:
    clus_info = pickle.load(handle)
clus_info
```

```python
# Loading dynamic signatures
with open('earm_signatures.pickle', 'rb') as handle:
    earm_signatures = pickle.load(handle)
# Obtaining signature of how the species 0 (the enzyme) is being consumend
sp37_sign_reactants = earm_signatures[37]['consumption']

# Initializing clustering object
clus = clustering.ClusterSequences(sp37_sign_reactants, unique_sequences=False)
clus.diss_matrix(n_jobs=4)

%matplotlib inline
# Clustering into two groups as 2 has the best Silhouette score
clus.Kmeans(n_clusters=10, n_jobs=4)
```

```python
# Plotting clustered signatures
pl = clustering.PlotSequences(clus)
pl.modal_plot()
```
